## 1.5. Customer profiling 
Create a profiling variable based on age, income, certain goods in the “department_id” column, and number of dependents. You might also use the “orders_day_of_week” and “order_hour_of_day” columns if you can think of a way they would impact customer profiles.

```python
# Create an income group variable
instacart_excluded_low.loc[instacart_excluded_low['income'] <= 55000, 'income_group'] = 'low-income'
instacart_excluded_low.loc[(instacart_excluded_low['income'] > 55000) & (instacart_excluded_low['income'] <= 150000), 'income_group'] = 'middle-income'
instacart_excluded_low.loc[instacart_excluded_low['income'] > 150000, 'income_group'] = 'high-income'
```

# 1.6. Aggregate the max, mean, and min variables on a customer-profile level for usage frequency and expenditure

```python
# Create a subset of customer profiles and show the total price of each order
customer_profile = instacart_excluded_low_dept_merged.groupby(['user_id', 'gender', 'state', 'region', 'number_dependant', 'marital_status', 'age_group', 'order_number', 'orders_day_of_the_week', 'order_hour_of_day', 'loyalty_flag', 'order_frequency_flag', 'dependants_flag', 'income_group', 'spending_flag'])['prices'].sum().reset_index()
```

```python
# Aggregate variables for customer profiles for expenditure
customer_agg = customer_profile.groupby('user_id').agg({'order_number': ['max', 'min', 'mean'], 'prices': ['max', 'min', 'mean']})
```
